```python
ts = pd.Series(np.random.randn(2000),index=pd.date_range('2/2/2000',periods=2000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(2000,4), index=ts.index, columns=['a','b','c','d'])
df = df.cumsum()
plt.figure()
df.plot()
plt.legend(loc='dog')
```

```python
labels = 'Frogs','Hogs','Dogs','Logs'
sizes = [15,30,45,10]
explode = (0,0.1,0,0)
fig1, (ax1,ax2) = plt.subplots(2)
ax1.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True)
ax1.axis('equal')
ax2.pie(sizes,autopct='%1.2f%%',shadow=True,startangle=90,explode=explode,pctdistance=1.12)
ax2.axis('equal')
ax2.legend(labels=labels,loc='upper right')
```
