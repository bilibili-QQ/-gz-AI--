```python
import cards_tool
'''
查询系统
'''
while True:
    cards_tool.show_menu()
    cmd_num = input("请选择执行的命令：")
    print("请选择执行的功能：%s" % cmd_num)
    if cmd_num == "1":
        print("新建名片")
    elif cmd_num == "2":
        print("显示全部")
    elif cmd_num == "3":
        print("查询名片")
    elif cmd_num == "0":
        print("退出系统")
        break
    else:
        print("请重新输入")
```

```python
r=int(input("请输入半径："))
PI = 3.14
def yuan():
    a=PI*r*r
    b=2*PI*r
    c=((4/3)*PI*r**3)
    d=4*PI*r**2
    print("面积:",a,"周长:",b,"体积:",c,"表面积:",d)
yuan()
```

```python
C = int(input("请输入摄氏度:"))
F = int(input("请输入华摄氏度:"))

a = C*1.8+32
b = (F-32)/1.8
print("摄氏度: {} 转换华摄氏度:{}".format(C,a))
print("华摄氏度: {} 转换摄氏度:{}".format(F,b))
```
